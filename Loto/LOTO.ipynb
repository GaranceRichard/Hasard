```python
class to_new_df:
    def __init__(self,data,day=None):
        """ create specific DataFrame return frequence with option by day"""
        if day is not None:
            data = data[data.jour_de_tirage.str.contains(day.upper())]
        self.size = data.shape[0] 
        
        new_df = pd.DataFrame(data.boule_1.value_counts()).sort_index()
        for x in range(2,6):
            item = "boule_" + str(x)
            b = data.eval(item).value_counts().sort_index()
            new_df[item] = b
        new_df = new_df.fillna(0)
        new_df["total"] = (new_df.boule_1+new_df.boule_2+new_df.boule_3+new_df.boule_4+new_df.boule_5)
        new_df["ratio"] = new_df.total/self.size
        
        self.base = data
        self.data = new_df
    
    def __repr__(self):
        print(self.data)
        
    def head(self):
        """Return head"""
        print(self.data.head())
        
    def get_data(self):
        """Get data"""
        return self.data

    def get_tirage(self):
        """Get tirage by date in cumsum"""
        tirages = pd.DataFrame([list(map(int,y.split("-"))) for y in [x.split("+")[0] for x in list(self.base.combinaison_gagnante_en_ordre_croissant)]],columns=["boule_"+str(x) for x in range(1,6)])
        tirages["date_de_tirage"] = df[["date_de_tirage"]]
        for x in range(1,50):
            tirages[str(x)]= np.max(np.where(tirages.isin([x]),1,0),axis=1)
            tirages[str(x)]= tirages[str(x)].cumsum()
            tirages[str(x)]= tirages[str(x)]/(tirages.index+1)
        return tirages
    
    def bestVal(self,limit=None):
        """Print best proba"""
        if limit is None:
            limit = 10
        data = self.data.sort_values(by="ratio", ascending=False)[:limit]
        print("Meilleurs numÃ©ros : {} \nAllant de {:,.2%} Ã  {:,.2%}".format(list(data.index), data.ratio.max(), data.ratio.min()))
        
        num_chance = self.base.numero_chance.value_counts().sort_values(ascending=False)/self.size
        index_chance = list(num_chance.index)[:3]
        proba_chance = list(num_chance)[:3]
        print("\nMeilleurs numÃ©ros chance : {} \nAllant de {:,.2%} Ã  {:,.2%}".format(index_chance,proba_chance[0],proba_chance[-1]))
    
    def worstVal(self,limit=None):
        """Print worst proba"""
        if limit is None:
            limit = 10
        data = self.data.sort_values(by="ratio", ascending=True)[:limit]
        print("Pires numÃ©ros : {} \nAllant de {:,.2%} Ã  {:,.2%}".format(list(data.index), data.ratio.min(), data.ratio.max()))
        
        num_chance = self.base.numero_chance.value_counts().sort_values(ascending=True)/self.size
        index_chance = list(num_chance.index)[:3]
        proba_chance = list(num_chance)[:3]
        print("\nPires numÃ©ros chance : {} \nAllant de {:,.2%} Ã  {:,.2%}".format(index_chance,proba_chance[0],proba_chance[-1]))
    
    def plot(self,descending=False):
        """Return graph from proba"""
        plt.figure(figsize=(20,10))
        if descending:
            data = self.data.sort_values(by="ratio", ascending=False).reset_index()
        else:
            data = self.data
        plt.bar(x=data.index, height=data.ratio)
        plt.show()

    def plot_tirage(self):
        """Return graph form tirages"""
        y = [str(x) for x in range(1,50)]
        b = self.get_tirage()
        b.plot(y=y, x='date_de_tirage', legend=False, figsize=(20,10))
        plt.show()
        
    
```

```python
a = to_new_df(df,"sam")
```
